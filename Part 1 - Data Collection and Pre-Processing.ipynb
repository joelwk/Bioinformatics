```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### **Select and retrieve bioactivity data for SARS coronavirus 3C-like proteinase (because we want single protein)**

We will assign the fifth entry to selected_target variable

```python
selected_target = targets.target_chembl_id[4]
selected_target
```

Retrieve only bioactivity data for coronavirus 3C-like proteinase (CHEMBL3927) that are reported as IC50 values in nM unit

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(3)
```

Finally we will save the resulting bioactivity data to a CSV file **bioactivity_data.csv**

```python
df.to_csv('bioactivity_data.csv', index=False)
```

## **Handling Missing Data**

If any compounds has mising value for the **standard_value** column then drop it

```python
df2 = df[df.standard_value.notna()]
```

```python
df2.head()
```

## **Labeling compounds as either being active, inactive or intermediate**

Bioactivity data is in the IC50 unit. Compounds having values less than 1,000 nM will be considered active while those greater than 10,000 nM will be considered inactive. Values between 1,000 and 10,000 will be considered intermediate

```python
bioactivity_class = []
for i in df2.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append('active')
    else:
        bioactivity_class.append('intermediate')
```

## **Create DataFrame for units of interest**

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[selection]
```

## **Iterate conical_smiles to a list**

```python
df3.head()
```

## **Add in bioactivity class**

Also, we want to name the bioactivity_class column otherwise it will display 0 as a header

```python
bioactivity_class = pd.Series(bioactivity_class, name='bioactivity_class')
df4 = pd.concat([df3, bioactivity_class], axis=1)
df4
```
